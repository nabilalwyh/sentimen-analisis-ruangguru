```python
# Import library google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort

# Import csv
import csv

# Import pandas untuk manipulasi dan analisis data
import pandas as pd


# Import library untuk mendownload hasil ulasan dalam file csv
from google.colab import files
```

# B. Scrapping Data

```python
# Mengambil hingga 25.000 ulasan paling relevan dari aplikasi dengan ID 'com.tiket.gits' di Google Play Store
scrapreview, continuation_token = reviews(
    'com.ruangguru.livestudents',            # ID aplikasi yang ingin diambil ulasannya
    lang='id',                   # Bahasa default ulasan yang diambil adalah Bahasa Indonesia
    country='id',                # Ulasan diambil dari pengguna yang berada di Indonesia
    sort=Sort.MOST_RELEVANT,     # Ulasan diurutkan berdasarkan relevansi tertinggi
    count=25000                  # Menentukan jumlah maksimal ulasan yang diambil yaitu 20.000
)
```

```python
# Menyimpan ulasan yang telah diambil ke dalam file CSV bernama 'ulasan_tiket_dot_com.csv'
with open('ulasan_ruangguru.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Judul kolom (header) di file CSV
    for review in scrapreview:
        writer.writerow([review['content']])  # Menuliskan isi ulasan ke dalam baris CSV satu per satu
```

```python
pd.options.mode.chained_assignment = None  # Mematikan peringatan tentang modifikasi data yang dilakukan secara chaining

# Mengonversi data ulasan (scrapreview) menjadi DataFrame
review_app = pd.DataFrame(scrapreview)

# Melihat beberapa baris awal dari DataFrame
review_app.head()

# Menyimpan data ulasan ke dalam file CSV tanpa menyertakan indeks
review_app.to_csv('ulasan_ruangguru.csv', index=False)
```

```python
# Menyimpan data frame baru dengan menghapus bari yang memiliki missing values
clean_df = review_app.dropna()

clean_df.shape
```

```python
#menghapus kolom yang tidak dibutuhkan
clean_df =clean_df.drop(['reviewId', 'userName', 'userImage','thumbsUpCount', 'reviewCreatedVersion', 'replyContent',
       'repliedAt', 'appVersion','at'], axis=1)
clean_df
```

```python
# Menghapus indeks
df_pemodelan = clean_df.reset_index(drop=True)
```

```python
# Mendownload dataset hasil scraping
files.download('ulasan_ruangguru.csv')

# Mendownload dataset hasil preprocessing data
df_pemodelan.to_csv('ulasan_pemodelan.csv',index=False)
files.download('ulasan_pemodelan.csv')

print("file berhasil didownload!")
```
